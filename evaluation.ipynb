```python
# 假设 pred 和 ground_truth 是您的预测和真实标签列表
pred = [['A', 'B', 'C', 'D', 'E'], ...]
ground_truth = [['E'], ...]

# 我们只关心每个预测的第一个标签
pred_top1 = [p[0] for p in pred]

# 计算准确率
accuracy = accuracy_score(ground_truth, pred_top1)

# 为了计算精确度、召回率和F1分数，我们需要将标签处理为二进制形式
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
ground_truth_binary = mlb.fit_transform(ground_truth)
pred_binary = mlb.transform(pred)

# 计算精确度、召回率和F1分数
precision = precision_score(ground_truth_binary, pred_binary, average='micro')
recall = recall_score(ground_truth_binary, pred_binary, average='micro')
f1 = f1_score(ground_truth_binary, pred_binary, average='micro')

# 打印结果
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
